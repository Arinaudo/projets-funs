```python
#packages 
import requests
from bs4 import BeautifulSoup
import sys
import pandas as pd
import numpy as np
import time
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#test scrapping 
url = 'https://www.lego.com/fr-fr/categories/price-under-20-eur?page='
response = requests.get(url)
if response.ok:
    print('Connexion réussie')
else: 
    sys.exit("BUG : Échec de connexion")

soup = BeautifulSoup(response.text, 'lxml')
#print(soup)
poupée=soup.find('ul',{'id':'product-listing-grid'})
#print(poupée) 
poupée1 = poupée.findAll('li', {'class': 'Grid_grid-item__FLJlN'})

tt_prix=[]
tt_nom=[]
tt_pièces=[]
for item in poupée1:
    prix=[]
    nom=[]
    pièces=[]
    nom_ = item.find('span', {'class': 'markup'})
    prix_ = item.find('span', {'class': 'ds-label-md-bold'})
    pièces_= item.find('span', {'data-test':'product-leaf-piece-count-label'})
    
    nom.append(nom_.text.strip() if nom_ else "Titre non disponible")
    prix.append(prix_.text.strip() if prix_ else "Prix non disponible")
    pièces.append(pièces_.text.strip() if pièces_ else "Pièces non disponible")
    tt_nom.append(nom_.text.strip() if nom_ else "Titre non disponible")
    tt_prix.append(prix_.text.strip() if prix_ else "Prix non disponible")
    tt_pièces.append(pièces_.text.strip() if pièces_ else "Pièces non disponible")
    
    prix=[x.replace("\xa0","") for x in prix]
    print(f"Produit:{nom}, Prix:{prix}, Pièces:{pièces}")
tt_prix=[x.replace("\xa0","")for x in tt_prix]
```
